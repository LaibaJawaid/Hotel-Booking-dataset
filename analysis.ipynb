```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df= pd.read_csv("hotel_booking.csv")
df
```

```python
df.info()
df.describe()
```

```python
df["desired_room_cat"] = np.where(df["reserved_room_type"] == df["assigned_room_type"],
                        "desired","undesired")
```

```python
print(df["desired_room_cat"].value_counts().sort_index())
```

```python
#Before Stratified Sampling
df["desired_room_cat"].value_counts().sort_index().plot.bar(rot=0, grid=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df["desired_room_cat"]):
    strata_train = df.loc[train_index]
    strata_test = df.loc[test_index]
```

```python
print(strata_train["desired_room_cat"].value_counts().sort_index())
strata_train
```

```python
print(strata_test["desired_room_cat"].value_counts().sort_index())
strata_test
```

```python
#After Stratified Sampling
strata_train["desired_room_cat"].value_counts().sort_index().plot.bar(rot = 0 ,grid = True )
```

```python
#After Stratified Sampling
strata_test["desired_room_cat"].value_counts().sort_index().plot.bar(rot = 0 ,grid = True )
```
